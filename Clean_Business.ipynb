```python
import json as js
import pandas as pd

bus = pd.read_json('./yelp_academic_dataset_business.json', lines=True)
bus = bus.dropna()
```

```python
discard = list()
for i in bus.index:
    if 'Restaurants' not in bus.loc[i].categories or bus.loc[i].is_open == 0:
        #print(bus.loc[i].categories)
        discard.append(i)
        
bus.drop(discard, inplace=True)
bus.drop('is_open', axis=1, inplace=True)
```

```python
f = open ('./Yelp_dataset_business_clean.json', 'w')

bus.to_json(path_or_buf=f)
f.close()
```
